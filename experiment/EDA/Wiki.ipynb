```python
import pandas as pd
data = pd.read_csv('./Data/Wiki/Wiki.csv')

```

```python
data=data.iloc[:,1:]
```

```python
len(data)
```

```python
l1=['Total']
l2,l3,l4,l5 = [],[],[],[]
for i in data.columns[1:]:
    if i.count('_') == 0:
        l2.append(i)
    elif i.count('_') == 1:
        l3.append(i)
    elif i.count('_') == 2:
        l4.append(i)
    else:
        l5.append(i)
```

```python
print(len(l1))
print(len(l2))
print(len(l3))
print(len(l4))
print(len(l5))
```

```python
import random
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

random.seed(10)
ll1 = random.sample(l2,3)
ll2 = random.sample(l3,3)
ll3 = random.sample(l4,3)
ll4 = random.sample(l5,3)
ll1.sort()
ll2.sort()
ll3.sort()
ll4.sort()

colors = ['red','green','blue']

x = [i for i in range(1,367)]
new_data = data

fig = plt.figure(figsize=(12, 9))

# 创建一个2x2的网格布局器
gs = gridspec.GridSpec(3, 2)

# 创建一个3x2的网格布局器
gs = gridspec.GridSpec(3, 2)

# 创建第一个子图，让它跨越两列
ax0 = fig.add_subplot(gs[0, :2])  # 这会让子图ax0占据第一行的前两列
ax0.set_title('Level_0')
# 在ax0上绘制图形
ax0.plot(x,new_data['Total'],label='T',color='black',linewidth=1)
ax0.legend()
ax0.set_xlabel('Time')

# 创建第二和第三个子图
ax1 = fig.add_subplot(gs[1, 0])
ax1.set_title('Level_1')
# 在ax1上绘制图形
for i in range(3):
    ax1.plot(x,new_data[ll1[i]],label=ll1[i],color=colors[i],linewidth=1)
ax1.legend()
ax1.set_xlabel('Time')

ax2 = fig.add_subplot(gs[1, 1])
ax2.set_title('Level_2')
# 在ax2上绘制图形
for i in range(3):
    ax2.plot(x,new_data[ll2[i]],label=ll2[i],color=colors[i],linewidth=1)
ax2.legend()
ax2.set_xlabel('Time')

# 创建第四和第五个子图
ax3 = fig.add_subplot(gs[2, 0])
ax3.set_title('Level_3')
# 在ax3上绘制图形
for i in range(3):
    ax3.plot(x,new_data[ll3[i]],label=ll3[i],color=colors[i],linewidth=1)
ax3.legend()
ax3.set_xlabel('Time')

ax4 = fig.add_subplot(gs[2, 1])
ax4.set_title('Level_4')
# 在ax4上绘制图形
for i in range(3):
    ax4.plot(x,new_data[ll4[i]],label=ll4[i],color=colors[i],linewidth=1)
ax4.legend()
ax4.set_xlabel('Time')

# 调整子图之间的间距
plt.tight_layout()

# 显示图表


plt.savefig('./Plot/Wiki/plot_data.png')

plt.show()
```
